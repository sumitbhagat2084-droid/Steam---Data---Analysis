# observation  : problem statment
1. Their are many games which are unknown and there specification is also null so now i will drop all the unknown rows.
2. In discount column there is % sign so we will remove it.
3. The release date column is object so we will change it to date and time
4. Price and original price column has free values so thats why that column is object so we will replce "free" with "0"
5. Discount column has "free" values so we will replace it with "100"
6. we will remove review column

```python
df.duplicated().sum()
```

```python
# Convert release_date to datetime
df['release_date'] = pd.to_datetime(df['release_date'], format='%d %b, %Y', errors='coerce')

# Check result
print(df['release_date'].head())
print(df['release_date'].dtype)
```

## observation 
- There are 153 duplicates so now i will drop them.

```python
df = df.drop_duplicates().reset_index(drop = True)
```

```python
df
```

```python
df.duplicated().sum()
```

# observation
- Now we have droped duplicates.
- And with this now we have solved our 1 problem statment unknown rows are also droped now.

#### Now we are solving our next problem statment which is of 2 %% sign.

```python
df['discount'] = df['discount'].str.replace(r'%%', '%', regex=True)
```

```python
df
```

# observation 
- Our 2 problem is also solved i have removed 1 % sign from the data set

### now we are converting free values into 0 from price and original price column

```python

# Replace "Free" with "0" in price and original_price columns
df["price"] = df["price"].astype(str).str.replace("Free", "0", case=False)
df["original_price"] = df["original_price"].astype(str).str.replace("Free", "0", case=False)

# Convert them to numeric (for calculations)
df["price"] = pd.to_numeric(df["price"], errors="coerce")
df["original_price"] = pd.to_numeric(df["original_price"], errors="coerce")
```

```python
df
```

# Observation
- our values are changed now

```python
df.to_csv("cleaned_steam_games1000.csv", index=False, encoding="utf-8-sig")
print("Data saved as steam_games.csv")
```

### Now we are replacing "free" values into 100% from discount column

```python
# Replace "Free" with "100%" in discount column
df["discount"] = df["discount"].astype(str).str.replace("Free", "100%", case=False)
df
```

# Observation
- our values are replaced now

### Now we are only keeping %values from review_rating column

```python
df["review_rating"] = df["review_rating"].astype(str).str.extract(r"(\d+%)")
df
```

# Observation 
- we have removed everything else rather than %

```python
df.head()
```

### we will remove review rating column

```python
df = df.drop(columns=["review_rating"], errors="ignore")
```
